```python
#Principais Imports
import pandas as pd
```

```python
df_2018 = pd.read_csv("http://dados.recife.pe.gov.br/dataset/45dbabee-0352-411a-b289-66fccde8942a/resource/5eaed1e8-aa7f-48d7-9512-638f80874870/download/sedec_chamados.csv", delimiter=";")
```

```python
df_2018
```

```python
num_colunas = len(df_2018.columns)
print("Número de colunas 2018:", num_colunas)
```

```python
# Usar o método melt para transformar as colunas em linhas
df_melted = pd.melt(df_2018)
print(df_melted)
```

```python
# Obter uma lista com os nomes de todas as colunas
lista_colunas = df_2018.columns.tolist()

# Exibir a lista de nomes de colunas
print(lista_colunas)
```

```python
# Configurar pandas para exibir todas as linhas
pd.set_option('display.max_rows', None)

# Criar um DataFrame com os nomes das colunas
colunas_nomes = pd.DataFrame({'Nomes de Colunas': df_2018.columns})

# Exibir o DataFrame de nomes de colunas
print(colunas_nomes)
```
